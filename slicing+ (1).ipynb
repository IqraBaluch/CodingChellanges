# slicing

```python
import numpy as np
arr=np.arange(1,25).reshape(12,2)
one=np.ones_like(arr)
one,arr
```

```python
one=np.ones((12,12))
one
```

```python
for i in range(len(one)):
    one[i::,::]=i
one
```

```python
i
```

```python
one
```

```python
one.astype("int32")
```

```python
j=len(one)-1
for i in range(len(one)):
    one[i::,::]=j
    j-=1
    
```

```python
one=np.ones((10,10))
one
```

```python
for i in range(len(one)):
    one[i::,::]=i
```

```python
print(one)
```

```python
for i in range(len(one)):
    one[:i:,:]=i
    print(one)
```

```python
j=len(one)-1
for i in range(len(one)):
    one[i::,::]=j
    j-=1
```
