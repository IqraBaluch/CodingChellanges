## Downloading data directly by link

```python
# Import requests package to request for data

import requests 
```

```python
# assign link of dataset to a vriable
url = "https://raw.githubusercontent.com/fivethirtyeight/data/master/nba-elo/nbaallelo.csv"

csv_path = "nba_all_elo.csv"

response = requests.get(url)
# to check if request was succesful
response.raise_for_status()
```

```python
#file handeling 
with open(csv_path, "wb") as file:
    file.write(response.content)
print("Download is ready")
```

Now the file (nba_all_elo.csv) is saved here/in current working directory.Use Pandas Python library to take a look at your data:

```python
import pandas as pd
nba = pd.read_csv("nba_all_elo.csv")
type(nba)
```

It seems that a team named "Lakers" played 6024 games, but only 5078 of those were played by the Los Angeles Lakers. Find out who the other "Lakers" team is

```python
nba.loc[nba['fran_id']=='Lakers', 'team_id'].value_counts()
```
