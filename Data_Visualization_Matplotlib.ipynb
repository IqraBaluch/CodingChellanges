## Customizing axis labels and adding titles

```python
#Import the matplotlib.pyplot submodule and name it plt
import matplotlib.pyplot as plt

# Create a Figure and an Axes with plt.subplots
fig, ax = plt.subplots()

ax.plot(austin['Date'], austin["DewPointAvgF"], color = 'r', marker= 'v', linestyle='--')
ax.plot(austin['Date'], austin["HumidityHighPercent"], color = 'yellow', marker= 'o', linestyle='--')

ax.set_xlabel("HumidityHighPercent_DewPointAvgF")
ax.set_ylabel("Precipitation")
ax.set_title("Weather patterns")
# Call the show function
plt.show()
```

# Creating small multiples with plt.subplots

```python
# Create a Figure and an array of subplots with 2 rows and 2 columns
fig, ax = plt.subplots(2, 2)

# Addressing the top left Axes as index 0, 0, plot month and Seattle precipitation


ax[0].plot(austin['Date'], austin["DewPointAvgF"], color = 'r', marker= 'v', linestyle='--')
ax[0].plot(austin['Date'], austin["HumidityHighPercent"], color = 'yellow', marker= 'o', linestyle='--')
ax[0].plot(austin["Date"], austin["	DewPointHighF"], color='r',marker= 's', linestyle='--')



ax[1].plot(austin['Date'], austin["DewPointAvgF"], color = 'r', marker= 'v', linestyle='--')
ax[1].plot(austin['Date'], austin["HumidityHighPercent"], color = 'yellow', marker= 'o', linestyle='--')
ax[1].plot(austin["Date"], austin["DewPointHighF"], color='r',marker= 's', linestyle='--')

plt.show()
```
